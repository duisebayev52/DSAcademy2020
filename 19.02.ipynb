```python
rows = 10
columns = 6
field = []
mines = [
    [1, 2],
    [2, 2],
    [5, 5],
    [7, 2],
    [8, 4],
]
for _ in range(rows):
    field.append([0] * columns)
#     field.append([])
#     for _ in range(columns):
#         field[-1].append(0)
    
```

```python
for mine in mines:
    field[mine[0]][mine[1]] = "*"
    if field[mine[0]-1][mine[1]-1] != "*" \
            and mine[0]-1 >= 0 and mine[1]-1 >= 0:
        field[mine[0]-1][mine[1]-1] += 1
    field[mine[0]-1][mine[1]] += 1
    field[mine[0]-1][mine[1]+1] += 1
    field[mine[0]+1][mine[1]-1] += 1
    field[mine[0]+1][mine[1]] += 1
    field[mine[0]+1][mine[1]+1] += 1
    field[mine[0]][mine[1]+1] += 1
    field[mine[0]][mine[1]-1] += 1
```
